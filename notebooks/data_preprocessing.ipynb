```python
import pandas as pd
import os

current_dir = os.getcwd()
data_path = os.path.join(current_dir, '..', 'data', 'raw', 'raw_data.csv')
data = pd.read_csv(data_path)
```

```python
# we will be dropping irrelevant columns (Comment made by VedantK)
data_model = data.drop(['RowNumber', 'CustomerId', 'Surname'],axis = 1)
```

```python
data_model.head()
```

```python
# Encoding Gender (cmt by vedant)
data_model['Gender'] = data_model['Gender'].map({'Male':1, 'Female':0})
```

```python
card_map = {'SILVER':1, 'GOLD':2, 'PLATINUM':3, 'DIAMOND':4}
data_model['Card_Type_Encoded'] = data_model['Card Type'].map(card_map)
data_model = data_model.drop('Card Type', axis=1)
```

```python
data_model = pd.get_dummies(data_model, columns=['Geography'])
```
